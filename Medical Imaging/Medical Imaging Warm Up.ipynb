```
import pydicom
img = pydicom.read_file('/content/gdrive/My Drive/bmed6590_medical_imaging/Mayo_medical_img.IMA')
print(img)
```

```
img = img.pixel_array
```

```
from matplotlib import pyplot as plt
plt.imshow(img - 1024.0, cmap='gray')
plt.show()
```

```
def normalize(img, lower=-160, upper=240):
  img = (img - 1024.0 - lower) / (upper - lower)
  img[img>1.0] = 1.0
  img[img<0.0] = 0.0
  return img
plt.imshow(normalize(img), cmap='gray')
plt.show()
```
